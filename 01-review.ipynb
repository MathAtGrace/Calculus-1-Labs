## Transformations

Given a function f(x), if

$y=m\cdot f[n \cdot (x-a)]+b,$

then f(x) will be:

- stretched vertically by a factor of |m|

- compressed horizontally by a factor of |n|

- displaced horizontally to the right by a units

- displaced vertically upward by b units

- If m is negative, f(x) will be inverted over the y-axis

- If n is negative, f(x) will be inverted over the x-axis

- The sign of a and b affects the direction of the function's displacement.

To apply this, try manipulating the function f(x) = 2*x in the cell below. Modify m, n, a, and b (and/or f(x) itself), then evaluate the code by clicking the 'Play' button or typing SHIFT-ENTER to see how the graph changes.

```python
m = -2
n = 1/10
a = 1
b = 2

def f(x):
    return 2*x

plot(m*f(n*(x-a))+b, x, -10, 10).show(ymin=-10, ymax=10)
```

## Trigonometry

Trigonometry is the math of triangles -- or as you have probably experienced, the math of triangles inscribed in circles. The following images from [Wikipedia](http://en.wikipedia.org/wiki/Unit_circle) are visual representations of the trigonometric functions and the unit circle, respectively.

<img src="trigfunctions.png">

This calculus tutorial requires no knowledge of versin or exsec, but you should be familiar with cosine, sine, and tangent, and have passing knowledge of cosecant, secant, and cotangent. Remember SOHCAHTOA (pronounced 'soak-uh-toe-uh'): Sine is Opposite over Hypotenuse, Cosine is Adjacent over Hypotenuse, and Tangent is Opposite over Adjacent.

<img src="unit_circle.png">

In the pictured unit circle, the coordinate pairs describe the cosine of the angle (comma) the sine of the angle. It shows that the cosine of pi over 3 is one-half, while the sine of pi over three is root three over two, and so forth. The units involving pi are known as radians: one would say, 'an angle of three pi over two radians', which is also equivalent to two hundred seventy degrees.

For this or any exploration of calculus, there are some basic trigonometric identities with which you should be familiar. In approximate order of importance, they are:

$\begin{align}\frac{\sin{x}}{\cos{x}}=\tan{x}\end{align}$

$\begin{align}\frac{1}{\cos{x}}=\sec{x}\end{align}$

$\begin{align}\frac{1}{\sin{x}}=\csc{x}\end{align}$

$\begin{align}\frac{1}{\tan{x}}=\cot{x}\end{align}$

$\begin{align}\sin^2{x} + \cos^2{x} = 1\end{align}$

$\begin{align}\sec^2{x} - \tan^2{x} = 1\end{align}$

$\begin{align}\csc^2{x} - \cot^2{x} = 1\end{align}$

$\begin{align}\sin{2x} = 2 \sin{x} \cos{x}\end{align}$

$\begin{align}\cos{2x} = \cos^2{x} - \sin^2{x} = 2 \cos^2{x} - 1=1 - 2 \sin^2{x}\end{align}$

$\begin{align}\tan{2x}=\frac{2 \tan{x}}{1 - \tan^2{x}}\end{align}$

To refresh your memory, here are some graphs of trigonometric functions:

```python
plot(cos(x),-2*pi,2*pi) + text('$f(x)=\cos{x}$', (pi,.6), fontsize=14, color='black')
```

```python
plot(sin(x),-2*pi,2*pi) + text('$f(x)=\sin{x}$', (1.5*pi,.6), fontsize=14, color='black')
```

```python
plot(tan(x),(x,-2*pi,2*pi),ymin=-10,ymax=10) + text(r'$f(x)=\tan{x}$', (pi,5), fontsize=14, color='black')
```

```python
plot(csc(x),(x,-2*pi,2*pi),ymin=-10,ymax=10) + text(r'$f(x)=\csc{x}$', (1.6*pi,5), fontsize=14, color='black')
```

```python
plot(sec(x),(x,-2*pi,2*pi),ymin=-10,ymax=10) + text(r'$f(x)=\sec{x}$', (pi,5), fontsize=14, color='black')
```

```python
plot(cot(x),(x,-2*pi,2*pi),ymin=-10,ymax=10) + text(r'$f(x)=\cot{x}$', (pi/2,5), fontsize=14, color='black')
```

The period of sine, cosine, secant, and cosecant is two pi, while the period of tangent and cotangent is pi. A function's period is the horizontal width of its repeating section. Since the sine function has a period of two pi, the interval [0, 2π] is identical to [2π, 4π]. Similarly, sine's interval [3π/8, 19π/8] is identical to [35π/8, 51π/8].
