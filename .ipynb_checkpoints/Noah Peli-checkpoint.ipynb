```python
#Exercise 2
g(x) = (2*x^2 - 6*x - 5)/(x-3)
a = solve( diff(g, x) == 0, x) #Tells us if we have roots
c = solve(g == 0, x) #x-intercepts
d = g(0) #y-intercepts
z = solve(1/g == 0, x) #vertical asymtotes
f = limit(g, x =+infinity), limit(g, x =-infinity) #Horizontal asymtotes
print("Has no roots or critical points", a, "x-intercepts", c,"y-intercepts", d,"vertical asymtotes", z,"horizontal asymtotes", f)
plot(g, x, -4, 4, detect_poles = True) .show(ymin = -20, ymax = 20)
```

```python
#Exerice 3
x = var('x')
g(x) = cos(x)*(e^(-x^2))
a, b = 1,5
n = 100
x_n =  [a + (b-a)/n*i for i in range(n+1)]
gx_n = [g(i).n() for i in x_n]
L = (b-a)/n * sum(gx_n[:-1])
R = (b-a)/n * sum(gx_n[1:])
print("Left Endpoints: ", L)
print("space")
print("Right Endpoints: ", R)
```
