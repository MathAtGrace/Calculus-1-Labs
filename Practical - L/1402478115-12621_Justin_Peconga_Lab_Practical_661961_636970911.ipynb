```python
def f(x): return sin(x)*e^-1*x^2
a, b = 1,5
n = 100
```

```python
print([i for i in range(n)])
print([i/2 + 1 for i in range(n)])
```

```python
x_n = [a + (b-a)/n*i for i in range(n+1)]
print(x_n)
```

```python
print(fx_n)
```

```python
print("Left endpoints: ", fx_n[:-1])
print("Right endpoints: ", fx_n[1:])
```

```python
L = (b-a)/n * sum(fx_n[:-1])
R = (b-a)/n * sum(fx_n[1:])
print("L: ", L)
print("R: ", R)
```
