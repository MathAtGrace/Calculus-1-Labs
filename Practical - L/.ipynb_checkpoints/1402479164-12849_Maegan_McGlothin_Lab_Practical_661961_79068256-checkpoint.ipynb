# Exercise 3

f(x) = sin(x)*e^-x^2

```python
def f(x): return sin(x)*e^-x^2
a, b = 1,5
n = 100
```

```python
print(list(range(n)))
print([i/2 + 1 for i in range (n)])
```

```python
x_n = [a + (b-a)/n*i for i in range (n+1)]
print(x_n)
```

```python
fx_n = [f(i).n() for i in x_n]
print(fx_n)
```

```python
print("Left endpoints: ", fx_n[:-1])
print("Right endpoints: ", fx_n[1:])
```

```python
L = (b-a)/n * sum(fx_n[:-1])
R = (b-a)/n * sum(fx_n[1:])
print("Left Approximation: ", L)
print("Right Approximation: ", R)
```

```python
P_f = plot(f(x), x, a, b)
P_L = sum([plot(fx_n[i], x, x_n[i], x_n[i+1], color = "red") 
          for i in range (n)])
P_R = sum([plot(fx_n[i+1], x, x_n[i], x_n[i+1], color = "green")
           for i in range(n)])
show(P_f + P_L + P_R)
```

For the function f(x), the Left Rieman sum is 0.135985460464793.
The Right Rieman sum for the function f(x) is 0.123603065438135.
