Since setting y = 0 and solving for x is giving me imaginary numbers, there are no x-intercepts. There is one y-intercept at (0, -5/3)

Local maximum and minimum

```python
show(solve(diff(g, x) > 0, x))
```

```python
show(solve(diff(g, x) < 0, x))
```

The above shows me the slope is positive whenever x < −1/2 * (√10) + 3 and negative when x is greater than the aformentioned expression. This means that −1/2 * (√10) + 3 is a local maximum. Similarly, the slope is positive when x > 1/2 * (√10) +3 and negative when x is less than the same expression. This means that 1/2 * (√10) + 3 is a local minimum.

# Problem 3

```python
def f(x): return sin(x) * e^((-x)^2)
a, b = 1, 5
n = 100
```

```python
print(list(range(n)))
print([i/2 + 1 for i in range(n)])
```

```python
x_n = [a + (b-a)/n*i for i in range(n+1)]
print(x_n)
```

```python
fx_n = [f(i).n() for i in x_n]
print(fx_n)
```

```python
print("Left endpoints: ", fx_n[:-1])
print("Right endpoints: ", fx_n[1:])
```

```python
L = (b-a)/n * sum(fx_n[:-1])
R = (b-a)/n * sum(fx_n[1:])
print("Left Approximation: ", L)
print("Right Approximation: ", R)
```

```python
P_f = plot(f(x), x, a, b)
P_L = sum([plot(fx_n[i], x, x_n[i], x_n[i+1], color = "red") for i in range (n)])
P_R = sum([plot(fx_n[i+1], x, x_n[i], x_n[i+1], color = "green") for i in range(n)])
show(P_f + P_L + P_R)
```

Above shows the left and right Rieman sums, as well as the graph. It's kind of amazing how the sums are so small (in magnitude) compared to the y-coordinates as x approaches 5. I guess it makes sense, though, as y is so tiny until x hits 4.5.
