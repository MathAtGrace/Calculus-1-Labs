```python
#Exercise 3

def f(x): return sin(x)*e^(-x^2)
a, b=1,5
n=100

#Find Partitions
print([i for i in range(n)])
print([i+1 for i in range(n)])
```

```python
#Create endpoints for partitions
x_n = [a+(b-a)/n*i for i in range(n+1)]
print(x_n)
```

```python
#Compute f(x) for the list
fx_n= [f(i).n() for i in x_n]
print(fx_n)
```

```python
#Find left endpoints 

print("Left endpoints:", fx_n[:-1])
```

```python
#Find right endpoints
print("Right endpoints:", fx_n[1:])
```

```python
#Sum Left Approximation

L=(b-a)/n*sum(fx_n[:-1])
print("Left Approximation:", L)
```

```python
#Sum right Approximation

R=(b-a)/n*sum(fx_n[:1])
print("Right Approximation:", R)
```
